```python
import pandas as pd

def calculate_text_statistics(df, text_columns):
    """
    Calculate the average number of characters and tokens per row for given text columns.

    Args:
        df (pd.DataFrame): The input DataFrame.
        text_columns (list): List of text columns to process.

    Returns:
        dict: A dictionary containing the average number of characters and tokens per row.
    """

    def estimate_tokens(text):
        """Estimate token count (approximation: 1 token ≈ 4 characters for English text)."""
        if pd.isna(text):  # Handle NaN values
            return 0
        if isinstance(text, dict):  # Convert dict to string if necessary
            text = str(text)
        return len(str(text).split())  # Ensure it's a string before splitting

    # Ensure all text columns are converted to strings before processing
    df[text_columns] = df[text_columns].astype(str)

    # Sum the number of characters for each row
    df["total_chars"] = df[text_columns].apply(lambda x: x.str.len()).sum(axis=1)

    # Calculate the estimated number of tokens per row
    df["total_tokens"] = df[text_columns].apply(lambda x: x.apply(estimate_tokens)).sum(axis=1)

    # Compute the averages
    avg_chars = df["total_chars"].mean()
    avg_tokens = df["total_tokens"].mean()

    return {
        "average_chars_per_row": avg_chars,
        "average_tokens_per_row": avg_tokens
    }

```

```python
import pandas as pd
import tiktoken

def calculate_text_statistics(df, text_columns):
    """
    Calculate the average number of characters and tokens per row for given text columns.
    It includes both a simple word-based token estimation and GPT tokenization using OpenAI's tokenizer.

    Args:
        df (pd.DataFrame): The input DataFrame.
        text_columns (list): List of text columns to process.

    Returns:
        dict: A dictionary containing the average number of characters, estimated tokens, and GPT tokens per row.
    """

    # Load OpenAI tokenizer (GPT-4o Mini uses "cl100k_base")
    enc = tiktoken.get_encoding("cl100k_base")  

    def estimate_tokens(text):
        """Estimate token count (approximation: 1 token ≈ 4 characters for English text)."""
        if pd.isna(text):  # Handle NaN values
            return 0
        if isinstance(text, dict):  # Convert dict to string if necessary
            text = str(text)
        return len(str(text).split())  # Word-based token approximation

    def gpt_tokenizer_count(text):
        """Count the actual token count using OpenAI's tokenizer."""
        if pd.isna(text):  # Handle NaN values
            return 0
        if isinstance(text, dict):  # Convert dict to string if necessary
            text = str(text)
        return len(enc.encode(text))  # Use GPT tokenizer

    # Ensure all text columns are converted to strings before processing
    df[text_columns] = df[text_columns].astype(str)

    # Sum the number of characters for each row
    df["total_chars"] = df[text_columns].apply(lambda x: x.str.len()).sum(axis=1)

    # Calculate the estimated number of tokens per row (word-based)
    df["total_tokens_estimate"] = df[text_columns].apply(lambda x: x.apply(estimate_tokens)).sum(axis=1)

    # Calculate the actual number of GPT tokens per row
    df["total_gpt_tokens"] = df[text_columns].apply(lambda x: x.apply(gpt_tokenizer_count)).sum(axis=1)

    # Compute the averages with rounding to 1 decimal place
    avg_chars = round(df["total_chars"].mean(), 1)
    avg_tokens_estimate = round(df["total_tokens_estimate"].mean(), 1)
    avg_gpt_tokens = round(df["total_gpt_tokens"].mean(), 1)

    return {
        "average_chars_per_row": avg_chars,
        "average_tokens_per_row (word-based)": avg_tokens_estimate,
        "average_gpt_tokens_per_row (OpenAI tokenizer)": avg_gpt_tokens
    }
```

```python
async def generate_signals_per_item(all_items_df, cols_for_prompt=['title', 'aspects', 'desc'], max_items=None,
                                    model_name='', outfile=None, debug=False, prompt_instructions_version="default",
                                    num_signals_per_item=5):
    """ Generate multiple signals per item based on the updated logic. """
    if model_name == 'gpt':
        signals_generator = SignalsGeneratorGPT(cols_for_prompt=cols_for_prompt,
                                                prompt_instructions_version=prompt_instructions_version,
                                                num_signals_per_item=num_signals_per_item)
    elif model_name == 'mistral':
        signals_generator = SignalsGeneratorMistral(cols_for_prompt=cols_for_prompt)
    else:
        raise ValueError(f"model_name: {model_name} is not supported")

    if max_items is None:
        max_items = len(all_items_df)

    if outfile is not None:
        with open(outfile, 'w') as file:
            file.write('')

    for ind, item in all_items_df[:max_items].iterrows():
        if debug:
            print(f"Index: {ind}, item: {item['item_id']}")

        # Generate signals (expecting multiple signals)
        gen_text = await signals_generator.generate_sig(item.to_frame().T, debug=debug)
        print(gen_text)

        if outfile is not None:
            with open(outfile, "a") as f:
                json.dump(gen_text, f, ensure_ascii=False)
                f.write("\n")
                

        # Extract multiple signals dynamically
        all_items_df.loc[ind, [f'gen_text_{model_name}'] + [f'gen_sig_{model_name}_{i+1}' for i in range(num_signals_per_item)] + [f'gen_exp_{model_name}_{i+1}' for i in range(num_signals_per_item)]] = \
            extract_sig_fix(gen_text, [f'gen_text_{model_name}', f'gen_sig_{model_name}'], num_signals_per_item)

    return all_items_df
```
