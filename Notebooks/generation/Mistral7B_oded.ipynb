```python
import torch
from ellement.transformers import AutoModelForCausalLM, AutoTokenizer
from ellement.peft import PeftConfig, PeftModel

base_model_mms_path = "mms://athena-mvp/Mistral-7B-Instruct-v02"
tokenizer = AutoTokenizer.from_pretrained(base_model_mms_path)
model = AutoModelForCausalLM.from_pretrained(base_model_mms_path, device_map="balanced", torch_dtype=torch.bfloat16) # On V100 switch to torch.float16


args = {'temperature': 0.8, 'max_new_tokens':256}
def mistral_gen(prompt_input):
    prompt_input_hf_format = [{"role": "user", "content": f"{prompt_input}"}]
    prompt = tokenizer.apply_chat_template(prompt_input_hf_format, tokenize=False, add_generation_prompt=True)
    inputs = tokenizer(prompt, return_tensors='pt').to('cuda')
    output = model.generate(**inputs, max_new_tokens=args['max_new_tokens'], do_sample=True, temperature=args['temperature'])
    return tokenizer.decode(output[0][len(inputs[0]):], skip_special_tokens=True)

mistral_gen("Why am I here?")
```

```python
import nltk
from nltk.tokenize import sent_tokenize
import re

# Ensure the necessary NLTK data is downloaded
# nltk.download('punkt')

def split_to_sentences_with_tokens(text):
    # Split the text into sentences
    sentences = sent_tokenize(text)
    
    # Find the splitting tokens
    split_tokens = re.findall(r'(?<=[.!?])\s+', text)
    
    # Ensure the number of tokens matches the number of splits
    if len(split_tokens) < len(sentences) - 1:
        split_tokens.extend([''] * (len(sentences) - 1 - len(split_tokens)))
    
    return sentences, split_tokens

# Example usage
text = "Hello world! This is a test. Let's split this text into sentences."
sentences, tokens = split_to_sentences_with_tokens(text)
print("Sentences:", sentences)
print("Tokens:", tokens)
```
