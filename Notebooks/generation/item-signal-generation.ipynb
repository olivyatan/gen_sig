```python
import gzip

file_path = '/data/shpx/notebooks/hroitman/ExpCnt/items.1.json.gz'
with gzip.open(file_path, "rt", encoding="utf-8") as f:
    lines = list(f.readlines())[::-1]
    all_items = [json.loads(line) for line in lines]

def item_type(item_title, types):
    for t in types:
        ts = t.split()
        ts = [w[:-1] if w.endswith('s') else w for w in ts if w != '&']
        num = 0
        for tss in ts:
            if tss in item_title:
                num += 1
        if num == len(ts):
            return t
    return "unknown"

types = [i["categ_lvl4_name"].lower() for i in ha_product_tips[0]]
for item in all_items:
    item["type"] =  item_type(item["title"].lower(), types)

all_items = [item for item in all_items if item["type"] != "unknown"]
print(f"num of items: {len(all_items)}")
```

```python
import warnings
warnings.filterwarnings("ignore")
import os
import sys
import base64
from bs4 import BeautifulSoup
import re
import pandas as pd

def html_to_plain(html: str):
    if html is None or len(html) == 0:
        return ""
    html = base64.b64decode(html).decode('utf-8')
    try:
        bs = BeautifulSoup(html, 'html.parser')
        plain = bs.get_text(separator=' ', strip=True)
        return re.sub(r"\s+", " ", plain)
    except:
        return ""

df = pd.read_parquet("home_app_item_desc")

df["desc"] = df["item_desc"].apply(lambda x: html_to_plain(x))

item_id_to_desc = dict()
for _, row in df.iterrows():
    item_id_to_desc[str(row["item_id"])] = row["desc"]
```

```python
import traceback
from tqdm import tqdm
import time
import pandas as pd

'''
Prompt to match product tips to a specific item
'''
item_tips_template =  '''
You are an expert salesperson that always excels in providing the best buying tips to buyers given a product of interest they wish to buy. 

Product details:
* Product title: {title}
* Product features: {aspects}
* Product description: {description}

Given the product's details above, your aim is first to identify which of the product features is the most appealing to make a convincing offer to the buyer. 
The feature you identify must appear in the product's details, with higher priority to a feature that is not part of the product's title. 
Think carefully why such a feature is the best to mention to the buyer and what would be your pitch strategy. 
As your answer, mention the product's feature and shortly generate one pitch sentence based on that feature. 
The feature's name should be informative, yet short, only three or four words. 
Use json answer format of: {{"feature": "short feature name", "sale pitch": "short sentence with feature-related sale pitch", "explanation": "your reasoning why this feature
is important"}}

'''

item_tips_prompt = PromptTemplate(
    input_variables=["tips", "title", "aspects", "description"],
    template=item_tips_template
)

item_tips_chain = LLMChain(llm=llm, prompt=item_tips_prompt)


verbose = True
debug = True
    

def run_item_tips_generation(item, product):
    try:
        description = item_id_to_desc[item['id']] if item['id'] in item_id_to_desc else ""
        item["tips"] = item_tips_chain.run({
            "tips": product["tips"],
            "title": item["title"],
            "aspects": item["aspects"],
            "description": description
        })
        item["product_tips"] = product["tips"]
        item["desc"] = description
    except Exception as e:
        if debug:
            print(e)
            traceback.print_exc()
        return False
    return True    
```

```python
import json
max_items = 100
items_to_save = list()
tips = ha_product_tips[0]
for product in tqdm(tips):
     
    for item in all_items:
        if item["type"] == product["categ_lvl4_name"].lower():
            if verbose or debug:
                print(f"item: {item['title']}")
            num_trials2 = 0
            while not run_item_tips_generation(item, product):
                sleep_time = 1 + 2 * num_trials2
                if debug:
                    print(f"sleeping {sleep_time} sec...")
                time.sleep(sleep_time)
                num_trials2 += 1
                if num_trials2 == 10:
                    break
            if verbose and 'tips' in item:
                if "Answer: " in item['tips']:
                    try:
                        item['tips'] = json.loads(item['tips'][8:])
                    except: 
                        continue
                print(f"item tips: {item['tips']}")  
                print(f"**********************************************")
            items_to_save.append(item)
            if len(items_to_save) == max_items:
                break
    if len(items_to_save) == max_items:
        break
        
print(f"total items: {len(items_to_save)}")
```

```python
from async_gpt import main_gpt_call, GPTCaller

class GPTCallerDescEdit(GPTCaller):
    def generate_prompt(self, aspects: str, chosen: str):
        return f'''
        The following is a description of an e-commerce product.

        Description: """
        {chosen}
        """

        Modify the above description by applying all of the following text editing instructions:
        1. First, identify all facts mentioned about the product in the above description. Lets call such facts: "description facts".
        2. Find facts in the "description facts" that also appear in the follow list of product features:
        """
        {aspects}
        """
        Lets call such facts "matching facts". 
        3. Remove from the description facts that are not part of the "matching facts". 
        4. Remove from the description any mentioned product features not part of the "matching facts".
        5. Remove from the description any text that includes "description facts" related to the product's condition, shippment, categories, delivery or seller contact.  
        6. Remove any text in the description that just repeat another one before it. 
        7. Check again that each fact of the "description facts" is indeed something mentioned in the above product's features.  
        8. Validate that the final description is synactically correct, grammatically correct and coherent.

        The final description may be empty as well. If it does, just state "empty description". 
        '''

pd.DataFrame(dict(chosen=['A very good item!', 'A very bad item!', 'Amzing listing!'], aspects=['Brand: Nike', 'Brand: Nike', 'Brand: Nike'])).to_json('try.jsonl', lines=True, orient='records')

await main_gpt_call(
    gpt_caller=GPTCallerDescEdit(), 
    data='try.jsonl', 
    output_path='out.jsonl', 
    n_tasks_slice=2,
    cols=['chosen', 'aspects']
)
```

```python
import re
import string
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize

stemmer = PorterStemmer()

def normalize_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Tokenize and stem each word
    tokens = word_tokenize(text)
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    # Join tokens back into a single string
    return ' '.join(stemmed_tokens)

def check_substring(T, s):
    # Normalize both texts
    normalized_T = normalize_text(T)
    normalized_s = normalize_text(s)
    # Use regex to check if normalized_s appears in normalized_T
    return re.search(r'\b' + re.escape(normalized_s) + r'\b', normalized_T) is not None

# Example usage
T = "This is a long text, with various punctuation! And some stemming."
s = "stem"
print(check_substring(T, s))  # Output: True
```
