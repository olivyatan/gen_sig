```python
import warnings
warnings.filterwarnings("ignore")
import os
import sys
import base64
from bs4 import BeautifulSoup
import re
import pandas as pd

def html_to_plain(html: str):
    if html is None or len(html) == 0:
        return ""
    html = base64.b64decode(html).decode('utf-8')
    try:
        bs = BeautifulSoup(html, 'html.parser')
        plain = bs.get_text(separator=' ', strip=True)
        return re.sub(r"\s+", " ", plain)
    except:
        return ""

```

```python
import base64
from bs4 import BeautifulSoup
def decode_base64_to_text(encoded_text):
    """Decodes a Base64 string and extracts readable text if it's HTML."""
    try:
        # Step 1: Decode Base64
        decoded_bytes = base64.b64decode(encoded_text)
        decoded_text = decoded_bytes.decode("utf-8", errors="ignore")

        # Step 2: Check if the decoded content is HTML
        if "<" in decoded_text and ">" in decoded_text:  # Quick HTML check
            soup = BeautifulSoup(decoded_text, "html.parser")
            decoded_text = soup.get_text(separator=" ")  # Extract plain text
        
        # Step 3: Remove excessive white spaces and return clean text
        return re.sub(r'\s+', ' ', decoded_text).strip()
    
    except Exception as e:
        return f"Error decoding text: {e}"

# Example Usage:
decoded_sentence = items_aspects_df_fashion.loc[0, 'item_desc']
plain_text = decode_base64_to_text(decoded_sentence)
print(plain_text)  # Output should be the human-reada
```
