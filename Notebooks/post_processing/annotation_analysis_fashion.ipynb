#### Join results of anotator pairs:

```python
import pandas as pd

def join_and_aggregate_pairs(annotation_dfs, annotator_pairs):
    """Join annotation DataFrames by pairs and return an aggregated DataFrame."""
    aggregated_df = pd.DataFrame()

    for annotator_1, annotator_2 in annotator_pairs:
        # Get the DataFrames for each annotator
        df1 = annotation_dfs.get(f"{annotator_1.lower()}_annotation_df")
        df2 = annotation_dfs.get(f"{annotator_2.lower()}_annotation_df")
        
        if df1 is not None and df2 is not None:
            # Ensure 'item_id' column is of the same type in both DataFrames
            df1['item_id'] = df1['item_id'].astype(str)
            df2['item_id'] = df2['item_id'].astype(str)

            # Rename columns in df2 to avoid conflicts
            df2 = df2.rename(columns={
                'annotator_name': 'annotator_name_2',
                'label': 'label_2',
                'reason': 'reason_2',
                'comment': 'comment_2'
            })
            
            # Join the two DataFrames on 'item_id'
            merged_df = pd.merge(df1, df2, on=['item_id','gen_sig_gpt'], how='inner')
            
            merged_df = merged_df.rename(columns={'title_x': 'title', 'aspects_x': 'aspects', 'desc_x': 'desc',
            'match_type_x': 'match_type', 'match_column_x': 'match_column'
            })

            # Select and reorder the required columns
            merged_df = merged_df[[
            'item_id', 'gen_sig_gpt', 'annotator_name', 'label', 'reason', 'comment',
            'annotator_name_2', 'label_2', 'reason_2', 'comment_2',
            'title', 'aspects', 'desc', 'match_type', 'match_column'
            ]]
            
            # Append the result to the aggregated DataFrame
            aggregated_df = pd.concat([aggregated_df, merged_df], ignore_index=True)

    return aggregated_df


# Example usage
annotator_pairs = [('Ortal', 'Matan'), ('Tal', 'Yonatan'), ('Guy', 'Yotam')]
output_df = join_and_aggregate_pairs(annotation_dfs, annotator_pairs)

```

```python
from sklearn.metrics import cohen_kappa_score


# List to store kappa scores
kappa_scores = []

# Compute Cohen's Kappa for each unique annotator pair
for pair in output_df[['annotator_name', 'annotator_name_2']].drop_duplicates().itertuples(index=False):
    # Filter data for this annotator pair
    pair_df = output_df[
        ((output_df['annotator_name'] == pair[0]) & (output_df['annotator_name_2'] == pair[1])) |
        ((output_df['annotator_name'] == pair[1]) & (output_df['annotator_name_2'] == pair[0]))
    ]
    
    if not pair_df.empty:
        # Drop NaN values to prevent errors
        valid_df = pair_df[['label', 'label_2']].dropna()

        if not valid_df.empty:
            kappa = cohen_kappa_score(valid_df['label'], valid_df['label_2'])
        else:
            kappa = None  # No valid data for this pair
    else:
        kappa = None  # No data for this pair

    kappa_scores.append({'annotator_pair': pair, 'Cohen_Kappa': kappa})

# Convert to DataFrame
kappa_df = pd.DataFrame(kappa_scores)

# Print Kappa scores per annotator pair
print(kappa_df)
```

```python
import pandas as pd

def join_with_final_disagree(output_df, df_final_disagree):
    """
    Joins output_df with df_final_disagree on ['item_id', 'gen_sig_gpt']
    and adds 'final_label', 'final_reason', 'final_comment' based on available values.

    - Ensures correct column naming even if suffixes (`_o`, `_d`) are added after join.
    - Prioritizes `df_final_disagree` values but falls back to `output_df` if missing.
    """

    # Ensure the join keys have matching dtypes
    output_df['item_id'] = output_df['item_id'].astype(str)
    output_df['gen_sig_gpt'] = output_df['gen_sig_gpt'].astype(str)
    df_final_disagree['item_id'] = df_final_disagree['item_id'].astype(str)
    df_final_disagree['gen_sig_gpt'] = df_final_disagree['gen_sig_gpt'].astype(str)

    # Perform a left join with suffixes
    joined_df = output_df.merge(df_final_disagree, on=['item_id', 'gen_sig_gpt'], how='left', suffixes=('_o', '_d'))

    # Creating final_label: Take from 'Final Label' if available, otherwise use 'label_o'
    joined_df['final_label'] = joined_df['final_label'].combine_first(joined_df['label_o'])

    # Creating final_reason: priority order → reason_d > reason_o > reason_2_o > reason_2_d
    joined_df['final_reason'] = joined_df[['final_reason','reason_d', 'reason_o', 'reason_2_o', 'reason_2_d']].bfill(axis=1).iloc[:, 0]

    # Creating final_comment: priority order → comment_d > comment_o > comment_2_o > comment_2_d
    joined_df['final_comment'] = joined_df[['final_comment','comment_d', 'comment_o', 'comment_2_o', 'comment_2_d']].bfill(axis=1).iloc[:, 0]

    # Define the required columns from output_df to retain
    required_columns = [
        'annotator_name', 'label', 'reason', 'comment', 'annotator_name_2', 'label_2',
        'reason_2', 'comment_2', 'title', 'aspects', 'desc', 'match_type', 'match_column'
    ]

    # Ensure correct column naming after merging (handle suffix issues)
    corrected_columns = []
    for col in required_columns:
        if f"{col}_o" in joined_df.columns:  # Keep original from output_df
            corrected_columns.append(f"{col}_o")
        elif f"{col}_d" in joined_df.columns:  # If missing, try from df_final_disagree
            corrected_columns.append(f"{col}_d")

    # Rename back to expected names (remove `_o` and `_d` suffixes)
    rename_dict = {col: col.replace('_o', '').replace('_d', '') for col in corrected_columns}
    joined_df = joined_df.rename(columns=rename_dict)

    # Final column selection: Put final_* columns after 'item_id' and 'gen_sig_gpt', followed by other columns from output_df
    selected_columns = ['item_id', 'gen_sig_gpt', 'final_label', 'final_reason', 'final_comment'] + list(rename_dict.values())

    return joined_df[selected_columns]

# Now apply the function
df_final_all_fashion_feb25 = join_with_final_disagree(output_df, df_final_disagree)
```
